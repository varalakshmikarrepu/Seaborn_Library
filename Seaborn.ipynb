```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note : Seaborn has some bulit-in datasets

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a these for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker')
```

```python
# house price dataset
import pandas as pd
from sklearn.datasets import fetch_california_housing

house_california = fetch_california_housing(as_frame=True)
house = house_california.frame
house.rename(columns={'MedHouseVal': 'price'}, inplace=True)

house.head()
```

Distribution plot

```python
import seaborn as sns
sns.distplot(house['price'])
```

Correlation:
postive
negative

Heat Map

```python
correlation = house.corr()
```

```python
import matplotlib.pyplot as plt
# constructing a heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f', annot=True,annot_kws={'size':8},cmap='Blues')
```
